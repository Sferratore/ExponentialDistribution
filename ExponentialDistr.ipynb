```python
from scipy.stats import expon
```

```python
import numpy as npy
npy.random.seed(0)
```

```python
import matplotlib.pyplot as pyplt
```

```python
x = npy.arange(0,5,0.1) # Represents an array of "time" values (0.1, 0.2, 0.3 ... 1.5, 1.6...). F.EX. Let's say that now the time unit for this problem is a year.
# So the array is representing 0.1 years, 0.2 years, ..., 1 year, 2 years...

y = expon.pdf(x, scale=1/2) # Calculates probability of two events happening at each x time, given that "scale" is the inverse of the media of events that happen in a defined time.
#F.EX. Now we are calculating for each yeartime from 0 to 5 how much there is possibility that two events happen in consecution in that time. We are using "scale" to indicate that the event occurs 2 times for standard time unit, which now is a year.

pyplt.plot(x, y)
```

You can see in the graph the probability density for each x point.

```python
y = expon.cdf(x, scale=1/2)  # Cumulative version of the function

pyplt.plot(x, y)
```

You can see in the graph that the probability that two consecutive events have happened at a little bit more than 0 years distance is around 0, at 1 year distance is around 0.3 etc.
